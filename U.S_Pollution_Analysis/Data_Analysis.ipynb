## There are various duplicates values present in US pollution data having same dates and as well as mean values.
## Thus its better to remove or approximate data. 

```python
air_pollution.columns
```

```python
air_pollution = air_pollution.drop(['Unnamed: 0','NO2 Units','O3 Units','SO2 Units','CO Units'],axis = 1)
```

```python
air_pollution
```

## There are missing values in SO2 AQI and CO AQI. In order to replace the missing values
## it can be replaced by mean having same State and Date 

```python
ap_clean = air_pollution[['State','Date Local','State Code','NO2 AQI','O3 AQI','SO2 AQI','CO AQI']]
ap_clean = ap_clean.dropna(axis='rows') 
ap_clean = ap_clean[ap_clean.State!='Country Of Mexico']
ap_clean['Date Local'] = pd.to_datetime(ap_clean['Date Local'])
ap_clean['Year'] = ap_clean['Date Local'].dt.year
ap_clean = ap_clean.groupby(['State','Date Local']).mean()
```

## Simple Linear Regression

```python
from sklearn.cross_validation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.4, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

### train set prediction

```python
plt.scatter(X_train, y_train, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
```

### Prediction Accuracy

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

rms = sqrt(mean_squared_error(y_test, y_pred))
rmsq = sqrt(mean_squared_error(y_train, regressor.predict(X_train)))
```

```python
rmsq
```

```python
import statsmodels.formula.api as sm
regressor_OLS = sm.OLS(endog = y, exog = X).fit()
regressor_OLS.summary()
```

## test set prediction

```python
plt.scatter(X_test, y_test, color = 'red')
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
```

## Polynomial Regression

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 5)
X_poly = poly_reg.fit_transform(X)
poly_reg.fit(X_poly, y)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

```python
plt.scatter(X, y, color = 'red')
plt.plot(X, lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
```

## prediction Accuracy

```python
from sklearn.metrics import mean_squared_error
from math import sqrt

rms = sqrt(mean_squared_error(y, lin_reg_2.predict(poly_reg.fit_transform(X))))
```
